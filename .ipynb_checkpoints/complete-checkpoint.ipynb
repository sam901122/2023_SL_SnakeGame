```python
# Importing libraries
from random import randint
import pygame
import time
```

```python
# Vairables
# Control variables
playing = True
pause = True

# Position variables
snake_positions = []
fruit_position = []
growth_position = [0, 0]

# Direction variables
move_up = False
move_down = False
move_left = False
move_right = True

# Constants
BOARD_LEN = 800
GRID_CNT = 20
GRID_STEP = BOARD_LEN / GRID_CNT

# Colors
grid_clrs = [(212, 191, 136), (230, 222, 185)]
border_clrs = [ (58, 73, 40),(75, 107, 60)]

# Speed
speed = 100
```

```python
# Functions
def project(pos):
	return [pos[0] * GRID_STEP, pos[1] * GRID_STEP]
```

```python
# 初始化 pygame
pygame.init()

# 建立視窗
window = pygame.display.set_mode((BOARD_LEN, BOARD_LEN))
windowRect = window.get_rect()
pygame.display.set_caption("Snake")

# 渲染視窗底色
cover = pygame.Surface(window.get_size())
cover = cover.convert()
cover.fill((250, 250, 250))
window.blit(cover, (0, 0))
for x in range(GRID_CNT):
    for y in range(GRID_CNT):
        rect = pygame.Rect(*project([x, y]), GRID_STEP, GRID_STEP)
        clr = grid_clrs[(x+y) % 2]
        if x == 0 or y == 0 or x == GRID_CNT - 1 or y == GRID_CNT - 1:
            clr = border_clrs[(x+y) % 2]
        pygame.draw.rect(cover, clr, rect)

window.blit(cover, (0, 0))
pygame.display.flip()
```

```python
# 載入圖片
head = pygame.image.load("./images/worm_head.png").convert_alpha()  # The head
head = pygame.transform.scale(head, (GRID_STEP * 1.1, GRID_STEP * 1.1))

body = pygame.image.load("./images/body.png").convert_alpha()  # The body
body = pygame.transform.scale(body, (GRID_STEP * 1.1, GRID_STEP * 1.1))

fruit = pygame.image.load("./images/fruit.png").convert_alpha()  # The fruit
fruit = pygame.transform.scale(fruit, (GRID_STEP * 1.2, GRID_STEP * 1.2))
```

```python
# Storing the variables in the list variables created before
snake_positions.append([2, 1])
snake_positions.append([1, 1])

# Giving random coordinates to the first fruit of the game
fruit_position = [randint(1, GRID_CNT - 2), randint(1, GRID_CNT - 2)]
```

```python
while playing:

    # Check events
    for event in pygame.event.get():

        # Quit
        if event.type == pygame.QUIT:
            playing = False

        # Press key
        if event.type == pygame.KEYDOWN:

            # Pause
            if event.key == pygame.K_SPACE:
                pause = not pause

            # Up
            elif event.key == pygame.K_UP:
                if move_down == False:
                    move_up = True
                    move_down = False
                    move_left = False
                    move_right = False

            # Down
            elif event.key == pygame.K_DOWN:
                if move_up == False:
                    move_up = False
                    move_down = True
                    move_left = False
                    move_right = False
            # Left
            elif event.key == pygame.K_LEFT:
                if move_right == False:
                    move_up = False
                    move_down = False
                    move_left = True
                    move_right = False

            # Right
            elif event.key == pygame.K_RIGHT:
                if move_left == False:
                    move_up = False
                    move_down = False
                    move_left = False
                    move_right = True

# Move
    if pause == False:
        # Body
        growth_position = snake_positions[-1]
        for i in range(len(snake_positions) - 1, 0, -1):
            snake_positions[i] = snake_positions[i - 1][:]

        # Head
        if move_up:
            snake_positions[0][1] -= 1
        elif move_down:
            snake_positions[0][1] += 1
        elif move_left:
            snake_positions[0][0] -= 1
        elif move_right:
            snake_positions[0][0] += 1

# Collision
# Collid itself
    for i in range(1, len(snake_positions)):
        if snake_positions[0] == snake_positions[i]:
            playing = False

    # Collid wall
    if snake_positions[0][0] >= GRID_CNT - 1:
        playing = False
    if snake_positions[0][0] <= 0:
        playing = False
    if snake_positions[0][1] >= GRID_CNT - 1:
        playing = False
    if snake_positions[0][1] <= 0:
        playing = False

# Eat Fruit
    if snake_positions[0] == fruit_position:

        # Generate fruit
        fruitInBody = True
        while fruitInBody:
            fruit_position = [randint(1, GRID_CNT - 2), randint(1, GRID_CNT - 2)]
            fruitInBody = False
            for i in range(len(snake_positions)):
                if snake_positions[i] == fruit_position:
                    fruitInBody = True

        snake_positions.append(growth_position[:])

        speed -= 2

    # Render
    # Cover
    window.blit(cover, (0, 0))
    # Snake body
    for i in range(1, len(snake_positions)):
        window.blit(body, (project(snake_positions[i])))
    # Snake head
    window.blit(head, project(snake_positions[0]))
    # Fruit
    window.blit(fruit, project(fruit_position))

    # Score
    font = pygame.font.SysFont(None, int(BOARD_LEN / 16))
    text = font.render("Score: " + str(len(snake_positions)), True, (0, 0, 0))
    window.blit(text, (BOARD_LEN * (3/4), 0))
    pygame.display.flip()

    # Delaying the game to make the snake move fluently
    time.sleep(speed / 1000)
```

```python
pygame.quit()
```
